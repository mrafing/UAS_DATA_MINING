```python
data2 = data.copy()
```

```python
#rata-rata
avg_age = data2['age'].mean()
avg_trtbps = data2['trtbps'].mean()
avg_chol = data2['chol'].mean()
avg_thalachh = data2['thalachh'].mean()
avg_oldpeak = data2['oldpeak'].mean()
```

```python
#now replace avg to columns 
data2['age'].replace({0:avg_age},inplace=True)
data2['trtbps'].replace({0:avg_trtbps},inplace=True)
data2['chol'].replace({0:avg_chol},inplace=True)
data2['thalachh'].replace({0:avg_thalachh},inplace=True)
data2['oldpeak'].replace({0:avg_oldpeak},inplace=True)
```

```python
from sklearn.model_selection import train_test_split

X = data2.drop('output',axis=1)
y = data2['output']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
# from sklearn.svm import SVC
# SVC_model = SVC()

from sklearn.neighbors import KNeighborsClassifier
KNN_model = KNeighborsClassifier()
```

```python
# SVC_model.fit(X_train,y_train)
KNN_model.fit(X_train, y_train)
```

```python
# svc_prediction = SVC_model.predict(X_test)
knn_prediction = KNN_model.predict(X_test)
```

```python
# from sklearn.metrics import classification_report, confusion_matrix
# print(confusion_matrix(y_test,svc_prediction))

from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, knn_prediction))
```

```python
# print(classification_report(y_test,svc_prediction))
print(classification_report(y_test, knn_prediction))
```
